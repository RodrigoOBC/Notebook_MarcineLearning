```
import torch
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```
np.random.seed(56)
X1,Y1 = make_classification(n_features=2, n_redundant=0, n_informative=1,n_clusters_per_class=1)
```

```
X1.shape
```

```
Y1.shape
```

```
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1)
p = X1[10]
plt.plot(p[0],p[1], marker='^', markersize=15)
```

```
def plotmodel(w1,w2,b):
  plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1)
  x = np.linspace(-2,4,50)
  y = (-w1*x -b)/ w2

  plt.axvline(0,-1,1, color='k', linewidth=1)
  plt.axhline(0,-2,4, color='k', linewidth=1)
  plt.plot(x,y)
  plt.grid(True)


w1 = -4.5
w2 = 5.5
b = -1.5
plotmodel(w1,w2,b)

p = (2,-2)
print(w1 * p[0] + w2 *p[1] + b)
```

```
def classify(ponto, w2,w1,b):
  ret =  w1 * ponto[0] + w2 *ponto[1] + b

  if ret >= 0:

    return (0,'purple')

  else:
    return (1, 'yellow')
```

```
p = (2,-2)

classse, color = classify(p,w1,w2,b)

print(f"cor = {color}, classe = {classse}")
plotmodel(w1,w2,b)
plt.plot(p[0],p[1], marker='^', markersize=15)
```

```
Positivo = 0

for k in range(len(X1)):
  classe, _ = classify(X1[k], w1,w2,b)
  if classe == Y1[k]:
    Positivo +=1

print(f"Acurácia: {Positivo/len(X1)}")
```
